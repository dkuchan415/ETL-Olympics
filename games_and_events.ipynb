```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
csv_file = "Resources/athlete_events.csv"
all_df = pd.read_csv(csv_file)
all_df.head()
```

```python
# Filtering all_df to represent all unique Olympic Games

games_df = all_df[["Year", "Season", "City"]]
games_df = games_df.drop_duplicates()
games_df = games_df.sort_values(["Year","Season"])
games_df = games_df.reset_index(drop=True)
games_df.index = np.arange(1, len(games_df)+1)
games_df
```

```python
# Filtering all_df to include all unique sporting events

events_df = all_df[["Sport","Event","Season"]]
events_df = events_df.drop_duplicates()
events_df = events_df.sort_values(["Season","Sport","Event"])
events_df = events_df.reset_index(drop=True)
events_df.index = np.arange(1, len(events_df)+1)

events_df
```

```python
rds_connection_string = "postgres:number12@localhost:5432/olympics_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
games_df.to_sql(name='games_df', con=engine, if_exists='append', index=False)
events_df.to_sql(name='events_df', con=engine, if_exists='append', index=False)
```
