```python
# Dependencies
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
# Name of the CSV file
athlete_events = 'Resources/athlete_events.csv'
```

```python
# The correct encoding must be used to read the CSV in pandas
athlete_eventsDF = pd.read_csv(athlete_events, encoding="ISO-8859-1")
```

```python
# Preview of the DataFrame
athlete_eventsDF.head()
```

```python
# Show athlete_event
athlete_event = athlete_eventsDF[["ID", "Age", "Sex", "Height", "Weight", "NOC", "Event", "Games", "Medal"]]
athlete_event = athlete_event.drop_duplicates()
```

```python
athlete_event
```

```python
# Show athlete_info
athlete_info = athlete_eventsDF[["ID", "Name", "Sex"]]
athlete_info = athlete_info.drop_duplicates()
athlete_info = athlete_info.set_index("ID")

```

```python
athlete_info
```

```python
rds_connection_string = "postgres:number12@localhost:5432/olympics_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
athlete_event.to_sql(name='athlete_event', con=engine, if_exists='append', index=False)
athlete_info.to_sql(name='athlete_info', con=engine, if_exists='append', index=False)
```
